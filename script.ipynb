```python
from dataset import DataModule
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
dm = DataModule('./data/PetImages/Cat', 2, 2)
```

```python
shapes = []
for f in dm.file_paths:
    with Image.open(f) as img:
        shapes.append(np.array(img.convert('RGB')).shape[:-1])
xs = [x for x,_ in shapes]
ys = [y for _,y in shapes]
```

```python
plt.scatter(xs, ys)
```

```python
len(list(filter(lambda x: x[0] >= 256 and x[1] >= 256, shapes)))
```

```python
aspect_ratios = [x / y for x,y in shapes]
plt.hist(aspect_ratios)
```
